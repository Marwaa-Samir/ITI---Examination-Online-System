```python
import pyodbc
import random
from datetime import datetime, timedelta

# Connect to the SQL Server database
connection_string = 'Driver={ODBC Driver 17 for SQL Server};Server=(local);Database=Examination System;Trusted_Connection=yes;'
connection = pyodbc.connect(connection_string)
cursor = connection.cursor()

# Exam Names
exam_names = [
    "Software Engineering",
    "Operating Systems",
    "Computer Organization and Architecture",
    "Object-Oriented Programming",
    "Data Structures and Algorithms",
    "Introduction to Programming",
    "Database Management Systems",
    "Web Development",
    "Cybersecurity",
    "Artificial Intelligence",
    "Software Testing",
    "Networking",
    "Mobile App Development",
    "Computer Graphics",
    "Human-Computer Interaction (HCI)",
    "Data Science",
    "Robotics",
    "Quantum Computing",
    "Augmented Reality (AR)",
    "Internet of Things (IoT)",
    "Blockchain Technology",
    "Cloud Computing"
]

try:
    # Generate fake data for exams
    for exam_name in exam_names:
        # Remove the word 'Exam' and trim spaces
        exam_name_without_exam = exam_name.replace("Exam", "").strip()

        # Find the corresponding course with the same name
        cursor.execute("SELECT Crs_Id FROM Course WHERE Crs_Name = ?", exam_name_without_exam)
        rows = cursor.fetchall()

        if rows:
            for row in rows:
                crs_id = row.Crs_Id
                num_exams = random.randint(1, 5)  # Random number of exams between 1 and 5

                for _ in range(num_exams):
                    # Generate a random date within the next 30 days
                    exam_date = datetime.now() + timedelta(days=random.randint(1, 30))

                    # Insert into the 'Exam' table
                    cursor.execute("INSERT INTO Exam (Exam_Name, Exam_Date, Crs_Id) VALUES (?, ?, ?)", exam_name, exam_date, crs_id)
        else:
            print(f"No matching course found for exam: {exam_name}. Skipping exam creation.")

    # Commit the changes
    connection.commit()
    print("Data inserted successfully.")
except Exception as e:
    print(f"Error: {e}")
finally:
    # Close the connection
    connection.close()
```

```python
import pyodbc
import random
from datetime import datetime, timedelta

# Connect to the SQL Server database
connection_string = 'Driver={ODBC Driver 17 for SQL Server};Server=(local);Database=Examination System;Trusted_Connection=yes;'
connection = pyodbc.connect(connection_string)
cursor = connection.cursor()

# Function to get unique non-repeating St_Id values from the Student table
def get_unique_student_ids():
    cursor.execute("SELECT TOP 250 St_Id FROM [dbo].[Student] ORDER BY NEWID()")
    return [row[0] for row in cursor.fetchall()]

# Function to generate random data for the Hiring table
def generate_hiring_data(num_records):
    hiring_data = []
    unique_student_ids = get_unique_student_ids()

    print("Number of Unique Student IDs:", len(unique_student_ids))
    print("Unique Student IDs:", unique_student_ids)

    # Fetching the range of Company_Id values from the Companies table
    cursor.execute("SELECT MIN(Company_Id), MAX(Company_Id) FROM [dbo].[Companies]")
    min_company_id, max_company_id = cursor.fetchone()

    # Fetching the range of Position_Id values from the Positions table
    cursor.execute("SELECT MIN(Position_Id), MAX(Position_Id) FROM [dbo].[Positions]")
    min_position_id, max_position_id = cursor.fetchone()

    for _ in range(num_records):
        # Check if there are enough unique student IDs
        if not unique_student_ids:
            break

        company_id = random.randint(min_company_id, max_company_id)  # Using the valid range
        position_id = random.randint(min_position_id, max_position_id)  # Using the valid range
        st_id = unique_student_ids.pop(0)  # Get a unique St_Id
        hire_date = datetime.now() - timedelta(days=random.randint(1, 365))  # Random hire date in the last year
        contract_length = random.randint(1, 12)  # Assuming contract length in months

        hiring_data.append((company_id, position_id, st_id, hire_date.date(), contract_length))

    return hiring_data

# Generate fake data for the Hiring table
num_records = 125  # Adjust the number of records as needed
hiring_data = generate_hiring_data(num_records)

# Insert fake data into the Hiring table
cursor.executemany("INSERT INTO [dbo].[Hiring] (Company_Id, Position_Id, St_Id, Hire_Date, ContractLength) "
                   "VALUES (?, ?, ?, ?, ?)", hiring_data)

# Commit the changes
connection.commit()

# Close the connection
connection.close()
```

```python
import pyodbc

# Connect to the SQL Server database
connection = pyodbc.connect('Driver={ODBC Driver 17 for SQL Server};'
                            'Server=(local);'
                            'Database=Examination System;'
                            'Trusted_Connection=yes;')

# Create a cursor
cursor = connection.cursor()

try:
    # Disable foreign key checks
    cursor.execute("EXEC sp_MSforeachtable 'ALTER TABLE ? NOCHECK CONSTRAINT ALL'")

    # Delete all data from all tables
    cursor.execute("EXEC sp_MSforeachtable 'DELETE FROM ?'")

    # Re-enable foreign key checks
    cursor.execute("EXEC sp_MSforeachtable 'ALTER TABLE ? WITH CHECK CHECK CONSTRAINT ALL'")

    # Commit the transaction
    connection.commit()

    print("All data deleted successfully.")
except Exception as e:
    # Rollback the transaction if an error occurs
    connection.rollback()
    print(f"Error: {e}")
finally:
    # Close the cursor and connection
    cursor.close()
    connection.close()
```
